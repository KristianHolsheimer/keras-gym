```python
%addpath --prepend ..
%reload skgym.utils
%reload skgym.value_functions.base skgym.value_functions.generic
%reload skgym.policies.base skgym.policies.value_based
%reload skgym.algorithms.base skgym.algorithms.td0

from skgym.value_functions.generic import GenericQTypeI, GenericQTypeII
from skgym.policies.value_based import ValueBasedPolicy
from skgym.algorithms.td0 import QLearning
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import FunctionTransformer, PolynomialFeatures
import time
import gym


# the Gym environment
env = gym.make('CartPole-v0')

num_episodes = 100
max_episode_steps = env._max_episode_steps

# # record video
# env = gym.wrappers.Monitor(env, os.path.join('data', 'video'))

# define sklearn model for approximating Q-function
regressor = SGDRegressor(eta0=0.05, power_t=0.0, fit_intercept=True)
transformer = FunctionTransformer(lambda x: np.hstack((x, x ** 2)), validate=False)
# transformer = PolynomialFeatures(degree=2, include_bias=False)

# define Q, its policy and update algorithm
q = GenericQTypeI(env, regressor, transformer)
policy = ValueBasedPolicy(q)
algo = QLearning(q, gamma=0.8)

# used for early stopping
num_consecutive_successes = 0


for episode in range(1, num_episodes + 1):
    last_episode = episode == num_episodes or num_consecutive_successes == 9
    
    # init
    s = env.reset()
    a = env.action_space.sample()
    
    if last_episode:
        env.render()
        epsilon = 0  # no more random exploration
    else:
        epsilon = 0.5 if episode < 10 else 0.01
    
    for t in range(1, max_episode_steps + 1):
        s_next, r, done, info = env.step(a)
        a_next = policy.epsilon_greedy(s_next, epsilon)
        
        # update or render
        if last_episode:
            env.render()
        else:
            algo.update(s, a, r, s_next)
        
        if done:
            if t == max_episode_steps:
                num_consecutive_successes += 1
                print(f"num_consecutive_successes = {num_consecutive_successes}")
            else:
                num_consecutive_successes = 0
                print(f"failed after {t} steps")
            break
    
        # prepare for next step
        s, a = s_next, a_next
        
    if last_episode:
        break


time.sleep(1)
env.close()
```
